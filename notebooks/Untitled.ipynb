# O-Notation Visualization
Used for presentation and report of opti-perf.

```python
import matplotlib.pyplot as plt
```

```python
x = range(1, 1000)
y = [i **2.807 for i in x]


# Plot the data
plt.plot(x, y)

# Add labels and title
plt.xlabel('')
plt.ylabel('')
plt.title('O(n^{2.327)')

# Display the plot
plt.show()
```
